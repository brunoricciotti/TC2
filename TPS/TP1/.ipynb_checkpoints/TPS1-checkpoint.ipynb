ec 1: \begin{equation}
(V_1 - V_x)SC= \frac{V_x}{R_3}
\end{equation}

entonces \begin{equation}
V_x = \frac{SCR_3*V_1}{1+SCR_3}
\end{equation} 

ec 2: \begin{equation} \frac{V_1-V_x}{R_1} = \frac{V_x-V_2}{R_2}\end{equation} 

reemplazando 2 en 1 y operando llegamos a:

\begin{equation} H(S) = \frac{S -\frac{R_2}{CR_1R_3}}{S + \frac{1}{CR_3}}\end{equation} 

$\omega_0 = \frac{1}{CR_3}$

\begin{equation} H(S) = \frac{S -\frac{R_2\omega_0}{R_1}}{S + \omega_0}\end{equation} 


Haciendo $S = jw$

\begin{equation} H(jw) = \frac{jw -\frac{R_2\omega_0}{R_1}}{jw + \omega_0}\end{equation} 

Modulo:

\begin{equation} |H(jw)| = \frac{\sqrt{w^{2} +(\frac{R_2\omega_0}{R_1}})^{2}}{\sqrt{w^{2} +\omega_0^{2}}}\end{equation} 

Fase:

\begin{equation}\theta = arctg(\frac{w}{-\frac{R2}{R1}.\omega_o})-arctg(\frac{w}{\omega_o})\end{equation}

Normalizando en frecuencia:

$S_n = \frac{S}{\omega_0}$ 

\begin{equation} H(S_n) = \frac{S_n \omega_0  -\frac{R_2\omega_0}{R_1}}{S_n \omega_0 + \omega_0}  = \frac{S_n  -\frac{R_2}{R_1}}{S_n + 1}\end{equation} 

Hacer esta normalizacion seria equivalente a multiplicar el valor de C por $\omega_0$. Y si ademas hacemos que $R_1 = R_2$ nos queda:

\begin{equation} H(S_n) = \frac{S_n  -1}{S_n + 1}\end{equation} 

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq

num = np.array([1, -1]) 
den = np.array([1, 1])

pretty_print_bicuad_omegayq(num,den)

tf = sig.TransferFunction(num,den)
    
plt.close('all')
analyze_sys(tf, 'TP1')
```
