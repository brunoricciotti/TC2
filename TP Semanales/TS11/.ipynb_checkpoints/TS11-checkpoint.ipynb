Lo primero que se debe hacer es eliminar esa Z1. Lo haremos de tal forma que en $Z_2 = Z - Z_1$, nos quede un cero en $s= -6$

Para que en $Z_2$ halla un cero $s= -6$, $Z_1$ debe ser $Z(-6) = \frac{8}{15}$



Entonces:

$Z_2(s) = \frac{(s^2+6s+8)}{(s^2+4s+3)} - \frac{8}{15}$

$Z_2(s) = \frac{7}{15}\frac{(s + \frac{16}{7})(s +6)}{(s^2+4s+3)}$

Como se ve, ahora tenemos un cero en $s= -6$

$Z_1 = 0.533$ es una resistencia.

Ahora hay que remover el primer tanque:

$Y_4 = Y_2 - \frac{s/R_1}{s + 1/R_1 C_1}$

$K_1 = \lim_{s\to -6}{Y_2 \frac{(s +6)}{s}} = \frac{75}{52}$

Sabiendo que $Y_{R_1 C_1}=\frac{s/R_1}{s + 1/R_1 C_1} = \frac{s\frac{52}{75}}{s + 6}$

$R_1 = \frac{52}{75}$ y $C_1 = \frac{25}{104}$

```python
from sympy import  *
import numpy as np
from sympy.abc import s
from IPython.display import display, Math
import math as m


#para que no aproxime
x1 = Rational('7/15')
x2 = Rational('58/15')
x3 = Rational('32/5')
x4 = Rational('75/52')

Y2 = (s**2 + 4*s + 3)/((x1)*s**2 + (x2)*s + x3)

Y2 = (simplify(Y2))

Y_rc1 = x4*s/(s+6)

Y4 = cancel(Y2-Y_rc1)

display(Y4)
```

Si ahora necesito un cero en $s = -3.5$, hago una remocion de forma que me genere ese cero.

Para que en $Z_6$ halla un cero $s= -3.5$, $Z_2$ debe ser $Z(-3.5) = \frac{884}{1005}$

```python
x7 = Rational('884/1005')

Z4 = 1/Y4

Z6 = cancel(Z4 - x7)

display(Z6)
```

$K_1 = \lim_{s\to -3.5}{Y_2 \frac{(s +3.5)}{s}} = \frac{4489}{4368}$

Sabiendo que $Y_{R_2 C_2}=\frac{s/R_2}{s + 1/R_2 C_2} = \frac{s\frac{4489}{4368}}{s + 3.5}$

$R_2 = \frac{4368}{4489}$ y $C_2 = \frac{4489}{15288}$

```python
x8 = Rational('4489/4368')
x9 = Rational('7/2')

Y6 = 1/Z6

Y_rc2 = x8*s/(s+ x9)

Y8 = Y6- Y_rc2

display(cancel(Y8))
```
