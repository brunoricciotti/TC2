```python
import sympy as sp
from sympy.abc import s
import numpy as np
import math as m
import matplotlib as mpl
import matplotlib.pyplot as plt
from splane import analyze_sys, pretty_print_bicuad_omegayq
from splane import pzmap, GroupDelay, bodePlot, modsq2mod_s
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt

T1 = 15/(s**3 + 6*s**2 + 15*s + 15)#transferencia de butter (S21(s))

T2 = 15/(-1*s**3 + 6*s**2 - 15*s + 15)#transferencia de butter (S21(-s))

T_cuad = sp.simplify(T1*T2)

#display(T_cuad)

S11_cuad = sp.simplify(1 - T_cuad)
display("|S_11|^2 = ",sp.simplify(S11_cuad))


coeffs_num=[1,0,-6,0,-45,0,0]
coeffs_den=[1,0,-6,0,-45,0,-255]

root = np.roots(coeffs_num)

display(root)

root = np.roots(coeffs_den)

display(root)

my_tf = TransferFunction( coeffs_num, coeffs_den )

pzmap(my_tf, 6)
```

Para obtener S11(s) usamos la funcion modsq2mod_s:

```python
S11 = modsq2mod_s(S11_cuad).evalf(4)

display("S11",S11)
```

Ahora despejo Z1 sabiendo que $S11 = \frac{Z_1 - R_0}{Z_1 + R_0}$

```python
Z1 = sp.simplify((S11+1)/(1-S11))

display("Z1",Z1)
```
