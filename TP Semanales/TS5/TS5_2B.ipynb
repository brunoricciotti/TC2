En el caso a, lo que podemos ver del grafico es que para $\omega=0$ la $T(s) = 0 dB$. 

Tambien se puede ver que para $\omega\rightarrow \infty$  $T(s) = -10 dB$. Esto solo puede ser posible si $k=-10 db = 0.316$

Entonces para $\omega=0$ la transferencia quedara:

$T(s) = \frac{k \omega_n^2}{\omega_p^2} =1$

Como la el denomindaor es el de un pasa-altos Butterworth, sabemos que $\omega_p = 1$. Finalmente no queda que $\omega_n = 1.779$. Esto se puede verificar en el grafico ya que para $\omega=\omega_n$ la $T(s) = 0$. 

```python
from sympy import  *
import numpy as np
from sympy.abc import s
from IPython.display import display, Math
from splane import pzmap, GroupDelay, bodePlot
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import math as m

wz = 1.779
wp = 1
Qp = float(1/sqrt(2))
k = 0.316


coeffs_num = [k*1, 0, k*(wz**2)]

display(coeffs_num)
coeffs_den = [1, (wp/Qp), wp**2]

root_num = np.roots(coeffs_num)
root_den = np.roots(coeffs_den)

display(root_den,root_num)

my_tf = TransferFunction(coeffs_num, coeffs_den )

pzmap(my_tf, 2) #S plane pole/zero plot


bodePlot(my_tf,2)
```

En el caso b, lo que podemos ver del grafico es que para $\omega=0$ y $\omega \rightarrow \infty$  la $T(s) = 0 dB$. 

Esto nos dice que $\omega_p = \omega_n = \omega_0$ 

Tambien se puede ver que para $\omega = \omega_0 = 1, T(1)=-6 dB = 0.5$

En $T(1) = \frac{Q_p}{Q_n}= -6 dB$, y como el enunciado dice que el denominador es el de un Butterworth, $Q_p = 1/\sqrt2$

Entonces $Q_n = \sqrt2$ 

Finalmente k =1 ya que no hay ganancia.

```python
wo = 1
Qz = float(sqrt(2))
Qp = float(1/sqrt(2))



coeffs_num = [1, (wo/Qz), wo**2]

display(coeffs_num)
coeffs_den = [1, (wo/Qp), wo**2]

root_num = np.roots(coeffs_num)
root_den = np.roots(coeffs_den)

display(root_den,root_num)

my_tf = TransferFunction(coeffs_num, coeffs_den )

pzmap(my_tf, 2) #S plane pole/zero plot


bodePlot(my_tf,2)
```
